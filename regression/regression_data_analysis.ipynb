# Regression Datasets Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set common styles
sns.set(style="whitegrid")
plt.rcParams.update({'font.size': 12})

regression_datasets_path = './datasets'
```

## 1. Fiat 500 Dataset Analysis

```python
fiat_df = pd.read_csv(os.path.join(regression_datasets_path, 'another-fiat-500-dataset-1538-rows.csv'))
```

### Fiat 500: Textual Analysis Output

```python
print("--- Fiat 500 Dataset Analysis ---")
print("\n### 1. Dataset Profile")
print(f"- Dimensions: {fiat_df.shape[0]} observations, {fiat_df.shape[1]} columns")
print(f"- Missing Values: {fiat_df.isnull().sum().sum()} total missing values")
print(f"- Duplicate Rows: {fiat_df.duplicated().sum()} duplicate rows")

print("\n### 2. Summary Statistics")
print(fiat_df.describe())

print("\n### 3. Correlation Matrix with Target (price)")
fiat_corr_price = fiat_df.select_dtypes(include=['float64', 'int64']).corr()['price'].sort_values(ascending=False)
print(fiat_corr_price)
```

### Fiat 500: Visualizations

```python
# 1. Correlation Heatmap
plt.figure(figsize=(12, 10))
fiat_corr = fiat_df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(fiat_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Fiat 500: Correlation Heatmap')
plt.show()

# 2. Target Variable Distribution (Price)
plt.figure(figsize=(10, 6))
sns.histplot(fiat_df['price'], kde=True, bins=30)
plt.title('Fiat 500: Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# 3. Price vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age_in_days', y='price', data=fiat_df, alpha=0.5)
plt.title('Fiat 500: Price vs. Age in Days')
plt.xlabel('Age in Days')
plt.ylabel('Price')
plt.show()

# 4. Price by Model
plt.figure(figsize=(12, 6))
sns.boxplot(x='model', y='price', data=fiat_df)
plt.title('Fiat 500: Price Distribution by Model')
plt.xlabel('Model')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

## 2. Laptop Price Dataset Analysis

```python
laptop_df = pd.read_csv(os.path.join(regression_datasets_path, 'laptop-price-prediction-dataset.csv'))
# Clean Ram and Weight columns
laptop_df['Ram'] = laptop_df['Ram'].str.replace('GB', '').astype(int)
laptop_df['Weight'] = laptop_df['Weight'].str.replace('kg', '').astype(float)
```

### Laptop Price: Textual Analysis Output

```python
print("--- Laptop Price Dataset Analysis ---")
print("\n### 1. Dataset Profile")
print(f"- Dimensions: {laptop_df.shape[0]} observations, {laptop_df.shape[1]} columns")
print(f"- Missing Values: {laptop_df.isnull().sum().sum()} total missing values")
print(f"- Duplicate Rows: {laptop_df.duplicated().sum()} duplicate rows")

print("\n### 2. Summary Statistics")
print(laptop_df.describe())

print("\n### 3. Correlation Matrix with Target (Price)")
laptop_corr_price = laptop_df.select_dtypes(include=['float64', 'int64']).corr()['Price'].sort_values(ascending=False)
print(laptop_corr_price)
```

### Laptop Price: Visualizations

```python
# 1. Target Variable Distribution (Price)
plt.figure(figsize=(10, 6))
sns.histplot(laptop_df['Price'], kde=True, bins=30)
plt.title('Laptop Price: Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# 2. Correlation Heatmap
plt.figure(figsize=(12, 10))
laptop_corr = laptop_df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(laptop_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Laptop Price: Correlation Heatmap')
plt.show()

# 3. Price vs. RAM
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ram', y='Price', data=laptop_df, alpha=0.5)
plt.title('Laptop Price vs. RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.show()

# 4. Price by Company
plt.figure(figsize=(14, 8))
sns.boxplot(x='Company', y='Price', data=laptop_df)
plt.title('Laptop Price Distribution by Company')
plt.xlabel('Company')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 3. Food Delivery Time Prediction Dataset Analysis

```python
food_df = pd.read_csv(os.path.join(regression_datasets_path, 'food-delivery-time-prediction.csv'))
```

### Food Delivery: Textual Analysis Output

```python
print("--- Food Delivery Dataset Analysis ---")
print("\n### 1. Dataset Profile")
print(f"- Dimensions: {food_df.shape[0]} observations, {food_df.shape[1]} columns")
print(f"- Missing Values: {food_df.isnull().sum().sum()} total missing values")
print(f"- Duplicate Rows: {food_df.duplicated().sum()} duplicate rows")

print("\n### 2. Summary Statistics")
print(food_df.describe())

print("\n### 3. Correlation Matrix with Target (Delivery_Time_min)")
food_corr_price = food_df.select_dtypes(include=['float64', 'int64']).corr()['Delivery_Time_min'].sort_values(ascending=False)
print(food_corr_price)
```

### Food Delivery: Visualizations

```python
# 1. Target Variable Distribution (Delivery Time)
plt.figure(figsize=(10, 6))
sns.histplot(food_df['Delivery_Time_min'], kde=True, bins=30)
plt.title('Food Delivery: Distribution of Delivery Time')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Frequency')
plt.show()

# 2. Correlation Heatmap
plt.figure(figsize=(12, 10))
food_corr = food_df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(food_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Food Delivery: Correlation Heatmap')
plt.show()

# 3. Delivery Time by Weather
plt.figure(figsize=(12, 6))
sns.boxplot(x='Weather', y='Delivery_Time_min', data=food_df)
plt.title('Food Delivery: Delivery Time by Weather')
plt.xlabel('Weather')
plt.ylabel('Delivery Time (minutes)')
plt.xticks(rotation=45)
plt.show()

# 4. Delivery Time by Traffic Level
plt.figure(figsize=(12, 6))
sns.boxplot(x='Traffic_Level', y='Delivery_Time_min', data=food_df)
plt.title('Food Delivery: Delivery Time by Traffic Level')
plt.xlabel('Traffic Level')
plt.ylabel('Delivery Time (minutes)')
plt.show()
```
